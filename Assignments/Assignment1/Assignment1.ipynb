## __Task 3__ 

```python
price_dict = {'Milk':50, 
              'Yoghurt':100, 
              'Paneer':200, 
              'Butter':250, 
              'Ghee':500, 
              'Butter Milk':400}
```

```python
price_dict.update({'Chicken':260})
print(price_dict)
```

```python
price_dict.update({'Paneer':240})
print(price_dict)
```

```python
if 'Yoghurt' in price_dict:
    price_dict.pop('Yoghurt')
    print(price_dict)
else:
    print("Product is not available.")
```

```python
total_price = sum(price_dict.values())
number_of_products = len(price_dict)
average_price = total_price / number_of_products

print(average_price)
```

```python
category_to_products = {'dairy':['Milk','Paneer','Butter'],
                        'Beverage': ['Butter Milk'],
                        'Non-Veg': ['Chicken']}
print(category_to_products)
```

```python
max_prod = len(category_to_products.values())
print(max_prod)
```

```python
for i in category_to_products:
    print(category_to_products[i])
```

```python
max_len = max(len(products) for products in category_to_products.values())
for category, products in category_to_products.items():
    if len(products) == max_len:
        print(products)
```
