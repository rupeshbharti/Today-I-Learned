## __Task 2__

```python
def factorial(n):
    if n < 0:
        print("Error: n cannot be negative")
        return None
    if n == 0 or n == 1:
        return 1

    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

```python
factorial(5)
```

```python
factorial(0)
```

```python
factorial(-3)
```

## __Task 6__

```python
def process_price(prices):

    discounted_prices = list(map(lambda p : p - p*0.10, prices))
    filtered_prices = list(filter(lambda p : p > 300, discounted_prices))

    return discounted_prices, filtered_prices

process_price([100, 500, 900, 50, 750])
```

## __Task 7__ : Mini Project

```python
# --- 1. Required Functions ---

def add_price(prices_list, price):
    """Adds the given price to the prices list."""
    prices_list.append(price)

def get_average_price(prices_list):
    """Returns the average price, or 0 if the list is empty."""
    if not prices_list:
        return 0
    return sum(prices_list) / len(prices_list)

def get_max_price(prices_list):
    """Returns the maximum price, or 0 if the list is empty."""
    if not prices_list:
        return 0
    return max(prices_list)


# --- 2. Main Program and Menu Loop ---

def main():
    # Initialize an empty list to store the prices
    prices = []
    
    while True:
        # Display the menu options
        print("\n--- Menu ---")
        print("1 -> Add price")
        print("2 -> Show average price")
        print("3 -> Show highest price")
        print("q -> Quit")
        
        # Take user input
        choice = input("Enter your choice: ").strip().lower()
        
        # Process the choice
        if choice == '1':
            try:
                new_price = float(input("Enter the price to add: "))
                if new_price < 0:
                    print("Price cannot be negative.")
                else:
                    add_price(prices, new_price)
                    print(f"Successfully added: ${new_price:.2f}")
            except ValueError:
                print("Invalid input. Please enter a valid numerical price.")
                
        elif choice == '2':
            avg = get_average_price(prices)
            print(f"The average price is: ${avg:.2f}")
            
        elif choice == '3':
            highest = get_max_price(prices)
            print(f"The highest price is: ${highest:.2f}")
            
        elif choice == 'q':
            print("Exiting the program. Goodbye!")
            break
            
        else:
            print("Invalid choice. Please enter 1, 2, 3, or q.")

# Run the program
if __name__ == "__main__":
    main()
```
