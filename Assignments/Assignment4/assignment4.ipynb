## __Task 1__

```python
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as f:
    for sale in sales:
        f.write(str(sale) + "\n")

print("Content of sales_data.txt")
with open("sales_data.txt", "r") as f:
    content = f.read()
    print(content)
```

## __Task 2__

```python
print("--- 1. Using .read() ---")
# 1. Read the entire file using .read() and print it.
with open("sales_data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
print("--- 2. Using .readline() ---")
# 2. Read the first line using .readline().
with open("sales_data.txt", "r") as file:
    first_line = file.readline()
    # .strip() removes trailing newline (\n) characters
    print(first_line.strip()) 
```

```python
print("\n--- 3. Using .readlines() ---")
# 3. Read all lines using .readlines() and convert them into a list of integers.
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    
    # Clean the strings and convert them to integers
    sales_integers = []
    for line in lines:
        cleaned_line = line.strip() # Removes \n
        if cleaned_line:            # Checks if the line is not empty
            int_value = int(cleaned_line)
            sales_integers.append(int_value)
            
    print("List of integers:", sales_integers)
```

## __Task 4__

```python
sales_list = []

with open("sales_data.txt", "r") as f:
    for line in f:
        cleaned_line = line.strip()

        if cleaned_line:
            sales_list.append(int(cleaned_line))

if sales_list:

    total_sales = sum(sales_list)
    highest_sale = max(sales_list)
    lowest_sale = min(sales_list)
    average_sale = total_sales / len(sales_list)

    # Print the results
    print(f"Total Sales: {total_sales}")
    print(f"Highest Sale: {highest_sale}")
    print(f"Lowest Sale: {lowest_sale}")
    print(f"Average Sale: {average_sale:.2f}")
else:
    print("The file is empty or contains no valid sales data.")
```

## __Task 5__

```python
# 1 & 2. Ask for 3 products and write them into products.txt
print("--- Enter Product Details ---")
with open("products.txt", "w") as file:
    for i in range(1, 4):
        name = input(f"Enter name for product {i}: ").strip()
        price = input(f"Enter price for product {i}: ").strip()
        
        # 3. Write in the format: ProductName | Price
        file.write(f"{name} | {price}\n")

print("\n--- Reading and Formatting File Content ---")

# 4. Read the file and print each line with proper formatting
with open("products.txt", "r") as file:
    for line in file:
        cleaned_line = line.strip()
        if cleaned_line:
            # Split the line by the pipe character
            parts = cleaned_line.split("|")
            
            # Extract and clean up spaces from each part
            product_name = parts[0].strip()
            product_price = parts[1].strip()
            
            # Print with clear visual formatting
            print(f"Product: {product_name:<15} Price: ${product_price}")
```

## __Task 7__

```python
# Create the initial dictionary
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}

# Ask the user for a discount percentage
discount_input = input("Enter discount percentage (e.g., 10 for 10%): ").strip()
discount_percent = float(discount_input)

# Variables to calculate summary metrics
total_discounted_price = 0
total_items = len(prices)

# Write to discount_report.txt
with open("discount_report.txt", "w") as file:
    # Process each product and calculate prices
    for product, original_price in prices.items():
        discount_amount = original_price * (discount_percent / 100)
        discounted_price = original_price - discount_amount
        total_discounted_price += discounted_price
        
        # Write format: Product | Original Price | Discounted Price
        file.write(f"{product} | {original_price} | {discounted_price:.2f}\n")
    
    # Extra (optional): Write summary at the bottom
    average_discounted = total_discounted_price / total_items
    file.write(f"Total Items: {total_items}\n")
    file.write(f"Average Discounted Price: {average_discounted:.2f}\n")

print("\n--- Reading discount_report.txt ---")

# Read the file and print it to the terminal
with open("discount_report.txt", "r") as file:
    content = file.read()
    print(content)
```
