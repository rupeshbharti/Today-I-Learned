## __Task 1__

```python
sales = [1200, 450, 980, 1500, 3000]

with open("sales_data.txt", "w") as f:
    for sale in sales:
        f.write(str(sale) + "\n")

print("Content of sales_data.txt")
with open("sales_data.txt", "r") as f:
    content = f.read()
    print(content)
```

## __Task 2__

```python
print("--- 1. Using .read() ---")
# 1. Read the entire file using .read() and print it.
with open("sales_data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
print("--- 2. Using .readline() ---")
# 2. Read the first line using .readline().
with open("sales_data.txt", "r") as file:
    first_line = file.readline()
    # .strip() removes trailing newline (\n) characters
    print(first_line.strip()) 
```

```python
print("\n--- 3. Using .readlines() ---")
# 3. Read all lines using .readlines() and convert them into a list of integers.
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    
    # Clean the strings and convert them to integers
    sales_integers = []
    for line in lines:
        cleaned_line = line.strip() # Removes \n
        if cleaned_line:            # Checks if the line is not empty
            int_value = int(cleaned_line)
            sales_integers.append(int_value)
            
    print("List of integers:", sales_integers)
```

## __Task 3__

```python
new_sales = [5000, 2500, 1700]

with open("sales_data.txt", "a") as f:
    for sale in new_sales:
        f.write(str(sale) + "\n")

print("--- Updated File Contents ---")
with open("sales_data.txt", "r") as file:
    contents = file.read()
    print(contents)

print("--- Extra Task ---")
with open("sales_data.txt", "r") as file:
    lines = file.readlines()
    total_lines = len(lines)
    print(f"Total number of lines in the file: {total_lines}")
```
