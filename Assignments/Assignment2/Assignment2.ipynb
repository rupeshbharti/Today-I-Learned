## __Task 1__

```python
while True:
    amount_str = input("Enter amount:")
    if amount_str.isdigit():
        order_amount = int(amount_str)
        break
    else:
        print("Invalid input. Please enter a numeric amount.")

if order_amount >= 2000:
  print("15 % discount")
elif 1500 <= order_amount < 2000:
  print("10 % discount")
elif 1000 <= order_amount < 1500:
  print("7 % discount")
else:
  print("No discount")
```

## __Task 2__

```python
orders = [1200, 2500, 800, 1750, 3000]

for order in orders:
  if order >= 2000:
    print("15 % discount")
  elif 1500 <= order < 2000:
    print("10 % discount")
  elif 1000 <= order < 1500:
    print("7 % discount")
  else:
    print("No discount")
```

```python
print("Order Amount | Discount % | Final Amount")
print("------------------------------------")
for order in orders:
  discount_percent = 0
  final_amount = order

  if order >= 2000:
    discount_percent = 15
    final_amount = order - (order * 0.15)
  elif 1500 <= order < 2000:
    discount_percent = 10
    final_amount = order - (order * 0.10)
  elif 1000 <= order < 1500:
    discount_percent = 7
    final_amount = order - (order * 0.07)

  print(f"{order:<12} | {discount_percent:<10} | {final_amount:<12.2f}")
```

```python
total_revenue = 0
for order in orders:
  final_amount = order
  if order >= 2000:
    final_amount = order - (order * 0.15)
  elif 1500 <= order < 2000:
    final_amount = order - (order * 0.10)
  elif 1000 <= order < 1500:
    final_amount = order - (order * 0.07)
  total_revenue += final_amount

print(f"Total revenue from all orders: {total_revenue:.2f}")
```

## __Task 3__

```python
order_amount = []

while True:
    print("\n1. Add order amount :")
    print("\n2. Calculate total revenue :")
    print("\nq. Quit")

    choice = input("Enter your choice: ")

    if choice == '1':
        amount_str = input("Enter the order amount: ")
        if amount_str.isdigit():
            amount = int(amount_str)
            order_amount.append(amount)
            print(f"Order amount {amount} added successfully.")
        else:
            print("Invalid input. Please enter a numeric amount.")
    elif choice == '2':
        if not order_amount:
            print("No orders to calculate revenue for. Please add orders first.")
        else:
            total_revenue = 0
            print("\nOrder Summary:")
            print("Order Amount | Discount % | Final Amount")
            print("------------------------------------")
            for order in order_amount:
                discount_percent = 0
                final_price = order # Default to no discount
                if order >= 2000:
                    discount_percent = 15
                    final_price = order - (order * 0.15)
                elif 1500 <= order < 2000:
                    discount_percent = 10
                    final_price = order - (order * 0.10)
                elif 1000 <= order < 1500:
                    discount_percent = 7
                    final_price = order - (order * 0.07)

                total_revenue += final_price
                print(f"{order:<12} | {discount_percent:<10} | {final_price:<12.2f}")
            print("------------------------------------")
            print(f"Total revenue from current orders: {total_revenue:.2f}")
    elif choice == 'q':
        print("Quiting program.")
        break
    else:
        print("Invalid choice. Please enter 1, 2, or q.")
        continue
```

## __Task 4__

```python
daily = [200,150,0,400,-1,50,300]

total_sales = 0
for sales in daily:
  if sales == -1:
    print("Corrupted data found. Stopping further processing.")
    break # Stop processing if corrupted data is found
  elif sales == 0:
    print("No sale today. Skipping this entry.")
    continue # Skip adding to total_sales if no sale
  else:
    print(f"Sale today: {sales}")
    total_sales += sales

print(f"Total sales from valid entries: {total_sales}")
```
