## Syntax:

class Class_name:

    def __init__(self, name, power, weapon, nationality):
        self.name = name
        self.power= power
        self.weapon= weapon
        self.nationality= nationality

    def introduction(self):
        print("my name is {self.name}")

## **Terminology** ->

### 1. Constructor ---> __init__ Method : 
The __init__() method is used to assign values to object properties, or to    perform operations that are necessary when the object is being created.
### 2. Self Parameter ---> 
The self parameter is a reference to the current instance of the class.
### 3. Class Properties ---> 
Properties are variables that belong to a class. They store data for each object created from the class.
### 4. Class Methods ---> 
Methods are functions that belong to a class. They define the behavior of objects created from the class.


## **Types of Constructor** ->

### 1. Default Constructor ---> 
It doesn't have the parameters inside it and it is only having self keyword inside it.
It assign a default value to all of the objects.

```python
class Dogs():

    def __init__(self):
        self.name="Tommy"
        self.breed="Labrador"

    def show_info(self):
        print(f"My pet name is {self.name} and his breed is {self.breed}")
```

```python
Dog1=Dogs()
```

```python
Dog1.name
```

```python
Dog1.show_info()
```

### 2. Parameterized Constructor --->
It consist parameters inside it so that every object have specific attributes associatd to it.

```python
class Avenger:
    def __init__(self, name, power, weapon, nationality ):
        self.name=name
        self.power=power
        self.weapon=weapon
        self.nationality=nationality

    def introduction(self):
        print(f"Hello, my name is {self.name}")

    def weapon_name(self):
        print(f"I am carrying the weapon {self.weapon}")
```

```python
Iron_man=Avenger("Iron Man","Genius & Billionaire","Iron Armor","America")

print(Iron_man.introduction())
print(Iron_man.weapon_name())
```

```python
Iron_man.weapon
```

## **Types of Methods** 

### Example :

```python
class BankAccount:
    # Class Variable (Shared by all accounts)
    interest_rate = 0.05  
    total_accounts_created = 0

    def __init__(self, owner_name, balance):
        # Instance Variables (Unique to each account)
        self.owner_name = owner_name
        self.balance = balance
        # Accessing class state to update global count
        BankAccount.total_accounts_created += 1

    # 1. INSTANCE METHOD
    # Automatically takes 'self' to access/modify object-specific data
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount}. New balance for {self.owner_name}: ${self.balance}")

    # 2. CLASS METHOD
    # Automatically takes 'cls' to access/modify class-level data
    @classmethod
    def change_interest_rate(cls, new_rate):
        cls.interest_rate = new_rate
        print(f"Bank-wide interest rate updated to: {cls.interest_rate * 100}%")

    # 3. STATIC METHOD
    # Completely independent; takes no hidden self/cls arguments
    @staticmethod
    def conversion_rate(usd_amount, rate=83.5):
        return usd_amount * rate
```

### 1. Instance Method : 


An instance method is a function that belongs to a specific object created from a class.
- Access: It can read and modify both the object's unique data (instance variables) and the overall class data.
- How it is called: It must be called through an object instance, not the class itself.
- Key feature: It automatically receives a reference to the specific object calling it (commonly named self in Python or this in Java/C++).
* Example: Checking the current balance of a specific user's bank account.

```python
account1 = BankAccount("Alice", 1000)
account2 = BankAccount("Bob", 500)

account1.deposit(250)  
account2.deposit(50)
```

### 2. Class Method : 

A class method is a function that belongs to the class itself, rather than to any individual object instance.
- Access: It can only access and modify class-level data (variables shared by all instances). It cannot access instance-specific data.
- How it is called: It can be called directly using the class name, without creating an object first.
- Key feature: It receives a reference to the class itself (commonly named cls in Python) as its first parameter.
* Example: Tracking the total number of bank accounts created across the entire bank system.

```python
BankAccount.change_interest_rate(0.06)

print(f"Total accounts tracking: {BankAccount.total_accounts_created}")
```

### 3. Static Method :

A static method is a utility function that lives inside a class namespace but behaves like a regular, independent function.
- Access: It cannot access or modify instance variables or class variables. It operates purely on the arguments passed directly to it.
- How it is called: It can be called directly using the class name, without creating an object.
- Key feature: It is grouped inside the class purely because it is logically related to the class topic, but it requires no internal class data to run.
* Example: A calculator function inside a Bank class that computes a compound interest rate based entirely on numbers you feed into it.

```python
rupee_value = BankAccount.conversion_rate(100)
print(f"$100 USD is roughly ₹{rupee_value} INR")
```
