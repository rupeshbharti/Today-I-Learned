# **Polymorphism**

It allows objects of different classes as if they are the object of same class & it enables reusability and makes program more flexible.

## __Types of Polymorphism__ :

### 1. Method Overriding (runtime polymorphism)

A child class provides a specific implementation of a method that has already been defined into the parent class. this allows dynamic selection at runtime.

```python
class Grandparent:

    def speak(self):
        print("My name is Rupesh")

class Parent:

    def speak(self):
        print("My name is Junior Rupesh")

class Child:

    def speak(self):
        print("My name is Junior Rupesh Junior")
```

```python
def attributes(objects):
    print(objects.speak())
```

```python
granpa=Grandparent()
```

```python
print(attributes(granpa))
```
