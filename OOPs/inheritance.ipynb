# __Inheritance__

It is fundamental concept of OOPs because of which child class is able to inherit the properties of parent class or attributes and functionality of parent class.

## __Types of Inheritance__

### 1. Single Inheritance

Parent Class properties ---> Child Class

```python
class Avenger:
    
    def __init__(self, name, power):
        self.name=name
        self.power=power

    def intro(self):
        print(f"My name is {self.name} ")
```

```python
class Ironman(Avenger):

    def __init__(self, name, power, weapon):
        super().__init__(name, power)
        self.weapon=weapon

    def weapon_intro(self):
        print(f"my power is {self.weapon}")
```

```python
Iron=Ironman("Tony Stark", "Genius Intellect", "Iron Suit")
```

```python
Iron.intro()
```

```python
Iron.weapon_intro()
```

### 2. Multilevel Inheritance

Grand Parent Class Properties ---> Parent Class Properties ---> Child Class 

```python
class Grand_parent:

    def __init__(self, height, baldness, face):
        self.height=height
        self.baldness=baldness
        self.face=face

    def grand_intro(self):
        print(f"Grand Parent height match {self.height} , baldness match {self.baldness} and face match {self.face}")
```

```python
class Parent(Grand_parent):

    def __init__(self, height, baldness, face, desease):
        super().__init__(height, baldness, face)
        self.desease=desease

    def parent_intro(self):
        print(f"Parent height match {self.height} , baldness match {self.baldness}, face match {self.face} and desease match {self.desease}")
```

```python
class Child(Parent):

    def __init__(self, height, baldness, face, desease, psycho):
        super().__init__(height, baldness, face, desease)
        self.psycho=psycho

    def child_intro(self):
        print(f"Child height match {self.height} , baldness match {self.baldness}, face match {self.face}, desease match {self.desease} and psycho {self.psycho}")
```

```python
child1=Child("Yes", "Yes", "No", "No", "Yes")
```

```python
child1.child_intro()
```

```python
child1.parent_intro()
```

```python
child1.grand_intro()
```

### 3. Multiple Inheritance

Parent_1 Class Properties + Parent_2 Class + Parent_3 Properties + ... ---> Child Class 

```python
class Parent1:

    def __init__(self, face, height):
        self.face=face
        self.height=height

    def p1_intro(self):
        print(f"Face = {self.face}, Heght = {self.height}")
```

```python
class Parent2:

    def __init__(self, desease, baldness):
        self.desease=desease
        self.baldness=baldness

    def p2_intro(self):
        print(f"Desease = {self.desease}, baldness = {self.baldness}")
```

```python
class Child(Parent1, Parent2):

    def __init__(self, face, height, desease, baldness, psycho):
        super().__init__(face, height, desease, baldness)
        self.psycho=psycho

    def ch_intro(self):
        print(f"Face = {self.face}, Heght = {self.height}, Desease = {self.desease}, baldness = {self.baldness}, psycho = {self.psycho}")
```

```python
child1=Child("Yes","No","Yes","No","Yes")
```
