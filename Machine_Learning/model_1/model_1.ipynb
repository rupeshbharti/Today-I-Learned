## Feature Engineering

```python
df['Total Cost'] = df['Tuition_USD'] + df['Rent_USD']
```

```python
df.head(5)
```

```python
df = df[['Country','City','University','Program',	'Level',	'Duration_Years','Living_Cost_Index','Visa_Fee_USD','Insurance_USD','Exchange_Rate','Total Cost']]
```

```python
df.head()
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['Country'] = le.fit_transform(df['Country'])
```

```python
df.head()
```

```python
df['City']=le.fit_transform(df['City'])
```

```python
df['University'] = le.fit_transform(df['University'])
```

```python
df['Program'] = le.fit_transform(df['Program'])
```

```python
df['Level'] = le.fit_transform(df['Level'])
```

```python
df.head(5)
```

## Separate Input Data & Output Data

```python
X = df.iloc[:,:9]
Y = df['Total Cost']
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
```

```python
X=s.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

## Stacking Regressor --> Ensemble Learning --> Combination of multiple Models of our own choice

```python
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
estimators = [('rf',RandomForestRegressor()),('br',BaggingRegressor())]
```

```python
stack = StackingRegressor(estimators=estimators,final_estimator=DecisionTreeRegressor())
```

```python
stack.fit(X_train,Y_train)
```

```python
y_predict = stack.predict(X_test)
```

```python
mean_absolute_error(Y_test,y_predict)
```

### Hyperparameter Tunning :
    - GridSearchCV
    - RandomSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {'n_estimators':[100,200],'max_depth' : [4,6,8]}

grid = GridSearchCV(RandomForestRegressor(),param_grid,cv=5)
```

```python
grid.fit(X_train,Y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {'n_estimators': [100,200],'max_depth': [4,6,8]}

random = RandomizedSearchCV(RandomForestRegressor(),param_distributions=param_grid,n_iter=10,cv=5)
```

```python
random.fit(X_train,Y_train)
```

```python
print(random.best_params_)
print(random.best_score_)
```
